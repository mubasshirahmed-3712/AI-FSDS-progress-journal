<h2 style="text-align:center;">üß∞ Section 2 ‚Äì Load Image and Convert to NumPy Array</h2>

In this section, we‚Äôll load an image using the <code>PIL.Image.open()</code> method and convert it to a <code>NumPy</code> array using <code>np.asarray()</code>.

This allows us to directly inspect and manipulate pixel-level data.

- ‚úÖ Understand image shape and channel structure
- ‚úÖ Prepare image for transformations and filtering
- ‚úÖ Visualize using <code>matplotlib.pyplot</code>

<hr>


```python
# ‚úÖ Import required libraries
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

# üì• Load image from input folder
img_path = '../input/3.jpg'
img = Image.open(img_path)

# ‚û§ Show original image
plt.imshow(img)
plt.axis('off')
plt.title("Original Image")
plt.show()
```

```python
# üîÑ Convert PIL image to NumPy array for pixel manipulation
car_arr = np.asarray(img)

# ‚û§ Print shape to understand dimensions (H x W x Channels)
print("‚úÖ NumPy array shape:", car_arr.shape)

# ‚û§ Optional: Preview the raw array (commented out if too large)
# print(car_arr)
```
