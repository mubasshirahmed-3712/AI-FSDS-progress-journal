```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings('ignore')

# Load the dataset
df = pd.read_csv(r'C:\Users\MUBASSHIR\OneDrive\Desktop\FSDS_TASKS\Projects\IrisViz-Matplotlib-Seaborn-Mastery\dataset\Iris.csv')

# Display first five rows
df.head()
```

<h2 style="text-align:center;">üìä Plot 1 ‚Äì Bar Plot: Count of Each Iris Species</h2>

This **bar plot** shows the total number of samples for each species in the Iris dataset.  
It helps us check for **balanced class distribution** before modeling.


```python
# Set Seaborn style for clean visuals
sns.set(style='whitegrid')

# ‚úÖ Plot: Bar plot of species count
sns.countplot(x='Species', data=df, palette='Set2')

# ‚û§ Customize labels and title
plt.title('Count of Iris Species', fontsize=13)
plt.xlabel('Species')
plt.ylabel('Sample Count')
plt.show()
```

<h2 style="text-align:center;">üìä Plot 2 ‚Äì Jointplot: Sepal Length vs Sepal Width</h2>

This **jointplot** combines a **scatter plot** with **histograms** to show the relationship between  
**Sepal Length** and **Sepal Width** across species.  
It helps visualize **distribution + correlation** between two key features.


```python
# ‚úÖ Jointplot: Sepal Length vs Sepal Width by species
sns.jointplot(data=df, x='SepalLengthCm', y='SepalWidthCm', hue='Species',
              kind='scatter', palette='Set1', height=6)

# ‚û§ Jointplot auto-displays; no need for plt.show()
```

<h2 style="text-align:center;">üìä Plot 3 ‚Äì Jointplot: Petal Length vs Petal Width</h2>

This **jointplot** shows how **Petal Length** correlates with **Petal Width** across different species.  
Highly useful for identifying **cluster separation** between species.


```python
# ‚úÖ Jointplot: Petal Length vs Petal Width by species
sns.jointplot(data=df, x='PetalLengthCm', y='PetalWidthCm', hue='Species',
              kind='scatter', palette='Set2', height=6)
```

<h2 style="text-align:center;">üìä Plot 4 ‚Äì Distribution of Sepal Length (Distplot)</h2>

This **distribution plot** shows the spread of **Sepal Length** across species.  
It helps visualize **central tendency**, **spread**, and **overlapping patterns**.


```python
# ‚úÖ Distribution Plot (Histogram + KDE) for Sepal Length
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x='SepalLengthCm', hue='Species', kde=True, palette='Set3', bins=15)

# ‚û§ Labels and title
plt.title('Distribution of Sepal Length by Species', fontsize=13)
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.show()
```

<h2 style="text-align:center;">üìä Plot 5 ‚Äì Boxplot of Sepal Length by Species</h2>

This **boxplot** shows the **median**, **quartiles**, and **outliers** of **Sepal Length** grouped by species.  
It‚Äôs a powerful tool to compare **distribution and variability** across species.


```python
# ‚úÖ Boxplot: Sepal Length grouped by Species
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='Species', y='SepalLengthCm', palette='Set2')

# ‚û§ Labels and title
plt.title('Boxplot of Sepal Length by Species', fontsize=13)
plt.xlabel('Species')
plt.ylabel('Sepal Length (cm)')
plt.show()
```

<h2 style="text-align:center;">üìä Plot 6 ‚Äì Heatmap: Correlation Matrix of Numerical Features</h2>

This **heatmap** displays the **correlation coefficients** between numerical features.  
It helps identify which features are **positively or negatively related** ‚Äî valuable for **feature selection**.


```python
# ‚úÖ Correlation Matrix
corr_matrix = df.drop('Id', axis=1).corr(numeric_only=True)

# ‚úÖ Heatmap of correlations
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)

# ‚û§ Title
plt.title('Feature Correlation Heatmap', fontsize=13)
plt.show()
```

<h2 style="text-align:center;">üìä Plot 7 ‚Äì KDE Plot: Sepal Width Density by Species</h2>

KDE plot (Kernel Density Estimate) shows **smooth distribution curves** for Sepal Width,  
helpful for comparing **data spread and peak values** across species.


```python
plt.figure(figsize=(8, 5))
sns.kdeplot(data=df, x='SepalWidthCm', hue='Species', fill=True, palette='pastel')
plt.title('KDE Plot: Sepal Width by Species', fontsize=13)
plt.xlabel('Sepal Width (cm)')
plt.ylabel('Density')
plt.show()
```

<h2 style="text-align:center;">üìä Plot 8 ‚Äì Pairplot of All Numerical Features</h2>

**Pairplot** displays **scatter plots + KDE plots** for all feature pairs,  
highlighting **relationships and separability** of species clusters.


```python
sns.pairplot(df.drop('Id', axis=1), hue='Species', palette='Set1', diag_kind='kde')
plt.suptitle('Pairplot of Features by Species', fontsize=14, y=1.02)
plt.show()
```

<h2 style="text-align:center;">üìä Plot 9 ‚Äì Violin Plot: Petal Width by Species</h2>

**Violin plots** combine boxplots and KDE,  
helping visualize **distribution + density + outliers**.

<h2 style="text-align:center;">üìä Plot 32 ‚Äì 3D Scatter Plot: Sepal & Petal Dimensions</h2>

Visualize **three dimensions** of Iris dataset using 3D scatter.


```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

for species in df['Species'].unique():
    subset = df[df['Species'] == species]
    ax.scatter(subset['SepalLengthCm'], subset['PetalLengthCm'], subset['PetalWidthCm'], label=species)

ax.set_xlabel('Sepal Length')
ax.set_ylabel('Petal Length')
ax.set_zlabel('Petal Width')
plt.title('3D Scatter Plot of Iris Dimensions')
plt.legend()
plt.show()
```

<h2 style="text-align:center;">üìä Plot 33 ‚Äì Radar Chart (Spider Plot) of Mean Feature per Species</h2>

Radar charts help compare **multidimensional mean values**  
for each species in a circular layout.


```python
# Calculate means
df_mean = df.groupby('Species').mean(numeric_only=True)

# Features to plot
labels = df_mean.columns.tolist()
angles = np.linspace(0, 2 * np.pi, len(labels), endpoint=False).tolist()
angles += angles[:1]  # Complete the circle

# Create Radar Chart
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))

for index, row in df_mean.iterrows():
    values = row.tolist()
    values += values[:1]
    ax.plot(angles, values, label=index)
    ax.fill(angles, values, alpha=0.1)

ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels)
plt.title('Radar Chart of Mean Feature per Species')
plt.legend(loc='upper right', bbox_to_anchor=(1.1, 1.1))
plt.show()
```
