<h2 style="text-align:center;">üîπ Section 1 ‚Äì Introduction to Matplotlib</h2>

Matplotlib is a powerful Python library for creating static, interactive, and animated visualizations.  
It is widely used in Data Science, Machine Learning, and scientific research.

We mostly use its `pyplot` module ‚Äî a collection of functions similar to MATLAB.

To get started, we must first install and import the necessary modules.


```python
# Importing required modules
import warnings
# Suppressing warnings
warnings.filterwarnings("ignore")
import matplotlib.pyplot as plt
import numpy as np


# Enabling inline plots (Jupyter only)
%matplotlib inline



# ‚û§ Output: Ready to use matplotlib and numpy
```

```python
# Creating basic line plot
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.plot(x, y)              # ‚û§ Line plot
plt.title("Simple Line Plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
# ‚û§ Output: Displays a line connecting the points (1,10), (2,20), etc.
```

<h2 style="text-align:center;">üìä Section 2 ‚Äì Plot Types in Matplotlib</h2>

In this section, we‚Äôll explore various plot types that are commonly used in data analysis and visualization with Matplotlib.  
Each plot is shown with clean, practical examples.

Plots Covered:
- Line Plot
- Scatter Plot
- Bar Chart
- Horizontal Bar Chart
- Stacked Bar Chart
- Error Bar Chart
- Histogram
- Pie Chart
- Box Plot
- Area Chart


```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, color='blue', linestyle='--', marker='o')
plt.title("üìà Line Plot ‚Äì Sine Wave")
plt.xlabel("X-axis")
plt.ylabel("Y = sin(x)")
plt.grid(True)
plt.show()
# ‚û§ Line plot of a sine wave with dashed lines and circular markers
```

```python
# scatter plot with random points
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y, color='green', marker='x')
plt.title("üîò Scatter Plot ‚Äì Random Points")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.show()
# ‚û§ Randomly scattered data points with cross markers
```

```python
# Bar Chart
names = ['Mubasshir', 'Farha', 'Ahmed', 'Tabassum']
scores = [88, 92, 75, 80]

plt.bar(names, scores, color='purple')
plt.title("üìä Bar Chart ‚Äì Student Scores")
plt.ylabel("Score")
plt.show()
# ‚û§ Simple vertical bar chart for categorical comparison
```

```python
# Horizontal Bar Chart

plt.barh(names, scores, color='orange')
plt.title("üìä Horizontal Bar Chart")
plt.xlabel("Score")
plt.show()
# ‚û§ Horizontal bars for same data
```

```python
# Stacked Bar Chart
subjects = ['Math', 'Science', 'English']
boys_scores = [75, 85, 90]
girls_scores = [80, 70, 88]

x = np.arange(len(subjects))
plt.bar(x, boys_scores, color='skyblue', label='Boys')
plt.bar(x, girls_scores, bottom=boys_scores, color='pink', label='Girls')

plt.xticks(x, subjects)
plt.title("üìä Stacked Bar Chart")
plt.ylabel("Scores")
plt.legend()
plt.show()
# ‚û§ Stacked bars showing scores of boys and girls across subjects
```

```python
# Error Bar Chart
x = np.arange(5)
y = np.random.normal(0, 1, 5)
errors = np.random.rand(5) * 0.3

plt.errorbar(x, y, yerr=errors, fmt='o', capsize=5, color='red')
plt.title("‚ö†Ô∏è Error Bar Chart")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.show()
# ‚û§ Shows uncertainty/error margins for data points
```

```python
# Histogram
data = np.random.randn(1000)

plt.hist(data, bins=30, color='steelblue', edgecolor='black')
plt.title("üìä Histogram ‚Äì Data Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
# ‚û§ Histogram showing distribution of 1000 random values
```

```python
# Pie Chart
labels = ['Python', 'Java', 'C++', 'SQL']
sizes = [45, 25, 15, 15]
colors = ['gold', 'lightgreen', 'skyblue', 'lightcoral']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.axis('equal')  # Equal aspect ratio ensures circular pie
plt.title("ü•ß Pie Chart ‚Äì Language Popularity")
plt.show()
# ‚û§ Pie chart with percentage labels and nice colors
```

```python
# Box Plot
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

plt.boxplot(data, vert=True, patch_artist=True)
plt.title("üì¶ Box Plot ‚Äì Data Spread")
plt.xlabel("Groups")
plt.ylabel("Value")
plt.grid(True)
plt.show()
# ‚û§ Box plot showing median, IQR, and outliers for each dataset
```

```python
# Area Chart
x = np.arange(1, 6)
y1 = np.array([1, 3, 4, 5, 7])
y2 = np.array([2, 2, 7, 3, 4])

plt.stackplot(x, y1, y2, labels=['Series A', 'Series B'], colors=['lightblue', 'lightgreen'])
plt.title("üèûÔ∏è Area Chart ‚Äì Stacked Series")
plt.xlabel("X-axis")
plt.ylabel("Cumulative Value")
plt.legend(loc='upper left')
plt.show()
# ‚û§ Stacked area chart showing cumulative contribution of series
```

<h2 style="text-align:center;">üé® Section 3 ‚Äì Customizing Plots in Matplotlib</h2>

In this section, we‚Äôll learn how to enhance the appearance and readability of our plots using:

- Titles & axis labels
- Legends
- Colors & styles
- Line width, markers, and grids
- Axis limits


```python
# 1. Titles, Axis Labels, and Grid
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.plot(x, y)
plt.title("üìå Line Chart Example")               # ‚û§ Add plot title
plt.xlabel("X-axis Label")                      # ‚û§ X-axis label
plt.ylabel("Y-axis Label")                      # ‚û§ Y-axis label
plt.grid(True)                                  # ‚û§ Show grid lines
plt.show()
```

```python
# 2. Adding Legends
x = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]
y2 = [1, 2, 3, 4]

plt.plot(x, y1, label='Squared', color='blue')
plt.plot(x, y2, label='Linear', color='orange')
plt.title("üìò Plot with Legends")
plt.legend()                  # ‚û§ Show legend
plt.show()
```

```python
#  3. Line Styles, Widths, and Markers
plt.plot(x, y1, linestyle='--', linewidth=2, marker='o', label='Dashed Line')
plt.plot(x, y2, linestyle='-.', linewidth=2, marker='x', label='Dash-dot Line')
plt.title("‚ú® Line Styles & Markers")
plt.legend()
plt.show()
```

```python
#  4. Color Customization
plt.plot(x, y1, color='green')     # Named color
plt.plot(x, y2, color='#FF5733')   # HEX color code
plt.title("üé® Custom Colors")
plt.show()
```

```python
# 5. Axis Limits and Tick Settings
plt.plot(x, y1)
plt.xlim(0, 5)                      # Set x-axis limits
plt.ylim(0, 20)                    # Set y-axis limits
plt.xticks([0, 1, 2, 3, 4, 5])     # Custom x-ticks
plt.yticks([0, 5, 10, 15, 20])     # Custom y-ticks
plt.title("üìê Axis Customization")
plt.show()
```

<h2 style="text-align:center;">üñºÔ∏è Section 4 ‚Äì Subplots and Multiple Figures</h2>

In this section, we will learn to:

- Plot multiple graphs in one figure
- Use `plt.subplot()` and `plt.subplots()`
- Customize layout spacing and figure size


```python
# ‚úÖ 1. Using plt.subplot() ‚Äì Manual Grid

import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]
y2 = [10, 20, 25, 30]

plt.figure(figsize=(10, 5))             # ‚û§ Create a new figure with custom size

plt.subplot(1, 2, 1)                    # ‚û§ Create 1st subplot in a 1x2 grid (1 row, 2 cols, plot 1)
plt.plot(x, y1, color='blue')          # ‚û§ Plot first dataset
plt.title("üìà Squared Line")           # ‚û§ Title for first subplot

plt.subplot(1, 2, 2)                    # ‚û§ Create 2nd subplot (1 row, 2 cols, plot 2)
plt.plot(x, y2, color='green')         # ‚û§ Plot second dataset
plt.title("üìä Linear Line")            # ‚û§ Title for second subplot

plt.tight_layout()                     # ‚û§ Adjust layout to prevent overlap
plt.show()
```

```python
# 2. Using plt.subplots() 

fig, ax = plt.subplots(2, 1, figsize=(6, 8))  # ‚û§ Create 2 subplots vertically (2 rows, 1 column)

ax[0].plot(x, y1, color='orange')            # ‚û§ Plot on first subplot
ax[0].set_title("üü† Squared Line")           # ‚û§ Title for first plot

ax[1].plot(x, y2, color='purple')            # ‚û§ Plot on second subplot
ax[1].set_title("üü£ Linear Line")            # ‚û§ Title for second plot

plt.tight_layout()                           # ‚û§ Prevent overlap between titles/axes
plt.show()
```

```python
#  3. Sharing Axes Between Plots
fig, ax = plt.subplots(1, 2, sharey=True)    # ‚û§ Create 2 side-by-side plots with shared y-axis

ax[0].bar(x, y1, color='skyblue')            # ‚û§ Bar plot on first subplot
ax[0].set_title("üìä Bar 1")                 # ‚û§ Title for first bar

ax[1].bar(x, y2, color='salmon')             # ‚û§ Bar plot on second subplot
ax[1].set_title("üìä Bar 2")                 # ‚û§ Title for second bar

plt.tight_layout()
plt.show()
```

```python
# 4. Creating Multiple Separate Figures
# ‚û§ First independent figure
plt.figure()
plt.plot(x, y1)
plt.title("Figure 1 ‚Äì Squared Line")         # ‚û§ Each figure is rendered separately

# ‚û§ Second independent figure
plt.figure()
plt.plot(x, y2)
plt.title("Figure 2 ‚Äì Linear Line")

plt.show()                                   # ‚û§ Show both figures
```

<h2 style="text-align:center;">üìä Section 5 ‚Äì Special Plot Types in Matplotlib</h2>

In this section, we'll explore advanced and practical plot types like:
- ‚úÖ Pie Charts
- ‚úÖ Box Plots
- ‚úÖ Area Charts
- ‚úÖ Contour Plots
- ‚úÖ Error Bar Plots
- ‚úÖ Stacked Bar Charts


```python
# 1. Pie Chart ‚Äì Data Composition Visualization

import matplotlib.pyplot as plt

# ‚û§ Sample data
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [45, 30, 15, 10]
colors = ['gold', 'lightcoral', 'lightskyblue', 'lightgreen']

# ‚û§ Explode the first slice (Python)
explode = (0.1, 0, 0, 0)

# ‚û§ Create pie chart
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title("üí° Pie Chart of Language Usage")
plt.axis('equal')  # ‚û§ Ensures the pie is a circle
plt.show()
```

```python
#  2. Box Plot ‚Äì Statistical Distribution Visualization
import numpy as np

# ‚û§ Generate random data
data = np.random.normal(100, 20, 200)  # mean=100, std=20, size=200

# ‚û§ Box plot
plt.boxplot(data)
plt.title("üì¶ Box Plot of Random Data")
plt.ylabel("Value")
plt.grid(True)
plt.show()
```

```python
#  3. Area Chart ‚Äì Cumulative Trends
# ‚û§ Sample data
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri']
values = [100, 120, 150, 170, 200]

# ‚û§ Area plot using fill_between
plt.plot(days, values, color='green')
plt.fill_between(days, values, color='lightgreen')
plt.title("üìà Area Chart ‚Äì Weekly Sales")
plt.ylabel("Sales")
plt.show()
```

```python
# 4. Contour Plot ‚Äì Level Lines for 3D Surface
import numpy as np

# ‚û§ Generate mesh grid
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(X**2 + Y**2)

# ‚û§ Contour plot
plt.contour(X, Y, Z, cmap='plasma')
plt.title("üó∫Ô∏è Contour Plot ‚Äì sin(x¬≤ + y¬≤)")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.colorbar(label='Z value')
plt.show()
```

```python
# 5. Error Bar Plot ‚Äì Displaying Variability
x = np.arange(1, 6)
y = np.exp(x)
errors = np.sqrt(y)  # ‚û§ Error = sqrt of values (just for demo)

# ‚û§ Plot with error bars
plt.errorbar(x, y, yerr=errors, fmt='o', color='teal', ecolor='orange', capsize=5)
plt.title("‚ùó Error Bar Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.show()
```

```python
# 6. Stacked Bar Chart ‚Äì Comparing Categories

import numpy as np

# ‚û§ Data
labels = ['A', 'B', 'C', 'D']
men = [20, 34, 30, 35]
women = [25, 32, 34, 20]

x = np.arange(len(labels))

# ‚û§ Plot stacked bars
plt.bar(x, men, label='Men', color='steelblue')
plt.bar(x, women, bottom=men, label='Women', color='lightpink')

plt.xticks(x, labels)
plt.ylabel("Scores")
plt.title("üë• Stacked Bar Chart ‚Äì Men vs Women")
plt.legend()
plt.show()
```

<h2 style="text-align:center;">üé® Section 6 ‚Äì Plot Customization and Styling in Matplotlib</h2>

This section focuses on improving the **visual aesthetics** and **usability** of your plots using:
- ‚úÖ Titles, Labels, Legends
- ‚úÖ Color Customization
- ‚úÖ Line Styles and Markers
- ‚úÖ Fonts and Ticks
- ‚úÖ Themes (Style Sheets)
- ‚úÖ Saving Plots to File


```python
#1. Adding Title, Axis Labels, and Grid
plt.plot([1, 2, 3, 4], [10, 20, 25, 30], marker='o')
plt.title("üìà Custom Line Plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.grid(True)
plt.show()
```

```python
#  2. Customizing Line Style, Width, and Marker
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.plot(x, y, color='purple', linestyle='--', linewidth=2, marker='s', markersize=8, markerfacecolor='yellow')
plt.title("üéØ Styled Line with Markers")
plt.grid(True)
plt.show()
```

```python
#  3. Using Fonts and Font Sizes
plt.plot(x, y)
plt.title("üìò Plot Title", fontsize=14, fontweight='bold', color='navy')
plt.xlabel("X Axis", fontsize=12)
plt.ylabel("Y Axis", fontsize=12)
plt.show()
```

```python
# 4. Using Matplotlib Style Sheets
plt.style.use('ggplot')  # ‚û§ Try 'seaborn', 'fivethirtyeight', 'bmh', etc.
plt.plot(x, y)
plt.title("üé® Styled Plot using ggplot")
plt.show()
```

```python
# 5. Saving Plots to File
plt.plot(x, y)
plt.title("üì§ Save This Plot")
plt.savefig("saved_plot.png", dpi=300, bbox_inches='tight')  # Save as image
plt.show()
```

```python
# 6. Adjusting Figure Size and Ticks
plt.figure(figsize=(8, 4))  # ‚û§ Width: 8, Height: 4 inches
plt.plot(x, y)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.title("üìè Custom Figure Size and Ticks")
plt.show()
```
